# Storytelling data visualisation on exchange rates

The dataset we'll use describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union.

If the exchange rate of the euro to the US dollar is 1.5, you get 1.5 US dollars if you pay 1.0 euro (one euro has more value than one US dollar at this exchange rate).

Daria Chemkaeva put together the data set and made it available on [Kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020)  — the data source is the European Central Bank. Note that the dataset gets regular updates — we downloaded it on January 2021.

Our focus in this project will be on the exchange rate between the euro and the American dollar.

# Introducing the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

# Data cleaning

```python
import re
new_col = []
for col in exchange_rates.columns:
    if re.search(r".(\\).", col):
        new_col.append(col.replace(col, 'Time'))
    elif re.match(r"\[.", col):
        new_col.append(col.replace('[', '').replace(' ]', '').replace(' ', '_'))
        
        
exchange_rates.columns = new_col
```

```python
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts().head()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'].value_counts().head()
```

```python
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

# Rolling mean

Now that we're finished cleaning the data, we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate.

```python
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

```python
fig, ax = plt.subplots(3, 2, figsize=(9,6))

rolling_means = [7, 30, 50, 100, 365]
n = 0
for i in range(3):
    for j in range(2):        
        if i == 0 and j == 0:
            ax[i,j].plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
            ax[i,j].set_title('Original values', weight='bold')

        else:
            ax[i,j].plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'].rolling(rolling_means[n]).mean())
            ax[i,j].set_title('Rolling_window: ' + str(rolling_means[n]), weight='bold')
            n += 1
        
plt.tight_layout() # Auto-adjusts the padding between subplots
plt.show()


# plt.figure(figsize=(9,6))

# plt.subplot(3,2,1)
# plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
# plt.title('Original values', weight='bold')

# for i, rolling_mean in zip([2, 3, 4, 5, 6],
#                            [7, 30, 50, 100, 365]):
#     plt.subplot(3,2,i)
#     plt.plot(euro_to_dollar['Time'],
#              euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
#     plt.title('Rolling Window:' + str(rolling_mean), weight='bold')
    
# plt.tight_layout() # Auto-adjusts the padding between subplots
# plt.show()
        
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar.tail()
```

# Coming up with an idea

Our goal is to create a storytelling data visualization using the data we have.

Here are a few story ideas for our data:

- We show how the euro-dollar rate has changed during the coronavirus pandemic. We can show the 2020 data and the 2016-2019 data as a baseline. We can use a line plot.
- We show how the euro-dollar rate changed during the 2007-2008 financial crisis. We can also show the data for 2006 and 2009 for comparison. We can use a line plot.
- We show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)). We can use a line plot.

# Coronavirus pandemic example

```python
import datetime as dt
import matplotlib.style as style

prior_pandemic = euro_to_dollar.copy()[euro_to_dollar['Time'].dt.year.between(2016, 2019)]
pandemic = euro_to_dollar.copy()[euro_to_dollar['Time'].dt.year >= 2020]

x0_pandemic = prior_pandemic['Time'].dt.date.iloc[0]

# Adding the FiveThirtyEight style
style.use('fivethirtyeight')

#Adding the subplots
fig, ax = plt.subplots(figsize=(8, 3))

ax.plot(prior_pandemic['Time'], prior_pandemic['rolling_mean'], linewidth=1, color='#B0C4DE')
ax.plot(pandemic['Time'], pandemic['rolling_mean'], linewidth=3, color='#e23d28')
ax.set_ylim(1, 1.3)
ax.set_yticks([1.05, 1.10, 1.15, 1.2])
ax.set_yticklabels(['1.05', '1.10','1.15', '1.2'], alpha=0.3, fontsize=11)
ax.set_xticklabels(['2016', '2017', '2018', '2019', '2020', '2021'], alpha=0.3, fontsize=11)

# Highlihting the peak of the pandemic
ax.axvspan(xmin='2020-12-15', xmax='2021-03-01', ymin=0.09, alpha=0.3, color='#D8BFD8')

#Adding a title and a subtitle
ax.text(x0_pandemic - dt.timedelta(300), 1.3, "Euro-USD rate peaked at 1.22 during 2020-2021's pandemic", weight='bold')
ax.text(x0_pandemic - dt.timedelta(300), 1.27, 'Euro-USD exchange rates between 2016 and 2021', size=12)

#Adding a signature
ax.text(x0_pandemic - dt.timedelta(300), 0.93, '©Tu_Fei' + ' '*115 + 'Source: European Central Bank',color = '#f0f0f0',
        backgroundcolor = '#800080', size=10)

plt.grid(alpha=0.5)
plt.show()
```

# Financial crisis example

```python
financial_crisis = euro_to_dollar.copy()[euro_to_dollar['Time'].dt.year.between(2006, 2009)]
financial_crisis_7_8 = euro_to_dollar.copy()[euro_to_dollar['Time'].dt.year.between(2007, 2008)]
x0 = financial_crisis['Time'].dt.date.iloc[0] - dt.timedelta(50)
xn = financial_crisis['Time'].dt.date.iloc[-1]

# Adding the FiveThirtyEight style
style.use('fivethirtyeight')

# Adding the plot
fig, ax = plt.subplots(figsize=(8,3))
ax.plot(financial_crisis['Time'], financial_crisis['rolling_mean'], linewidth=1, color='#B0C4DE')
ax.plot(financial_crisis_7_8['Time'], financial_crisis_7_8['rolling_mean'], linewidth=3, color='#e23d28')
ax.set_xlim(x0, xn)

# Highlihting the peak of the crisis
ax.axvspan(xmin='2008-04-01', xmax='2008-09-01', ymin=0.09, alpha=0.3, color='#D8BFD8')

#Adding separate tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

x = x0

for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, 1.13, year, alpha=0.5, fontsize=11)
    x += dt.timedelta(365)
    

y = 1.193
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(x0-dt.timedelta(90), y, rate, alpha=0.5, fontsize=11)
    y += 0.1
    
#Adding a title and a subtitle
ax.text(x0-dt.timedelta(90), 1.67, "Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis", weight='bold')
ax.text(x0-dt.timedelta(90), 1.63, 'Euro-USD exchange rates between 2006 and 2010', size=12)

#Adding a signature
ax.text(x0-dt.timedelta(90), 1.07, '©Tu_Fei' + ' '*110 + 'Source: European Central Bank',color = '#f0f0f0',
        backgroundcolor = '#800080', size=10)
    
#Add some transparency to the grid
ax.grid(alpha=0.5)
plt.show()
```

# The three US presidencies example

```python
bush_obama_trump = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2021)]
bush = bush_obama_trump.copy()[bush_obama_trump['Time'].dt.year < 2009]
obama = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >= 2009) & (bush_obama_trump['Time'].dt.year < 2017)]
trump = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >= 2017) & (bush_obama_trump['Time'].dt.year < 2021)]
```


Below, you'll notice we used matplotlib's functional approach to build the graphs. We use this approach because it offers more flexibility in arranging the subplots:

- We first build three of the graphs on a 2-by-3 grid (this grid should have six subplots, but we only build three; the bottom row remains empty).
- We then build only the bottom graph of a 2-by-1 grid (this grid should have two subplots; the top row remains empty).
- The two grids are merged, and we end up with three graphs on the top row and one graph on the bottom row.

```python
# Adding the FiveThirtyEight style
style.use('fivethirtyeight')

#Adding the subplots
plt.figure(figsize=(12, 6))
ax1 = plt.subplot(231)
ax2 = plt.subplot(232)
ax3 = plt.subplot(233)
ax4 = plt.subplot(212)
axes = [ax1, ax2, ax3, ax4]

#Changes to all subplots
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2','1.4', '1.6'], alpha=0.3)
    ax.grid(alpha=0.5)


def plot_presidencies(ax, name, period, x_data, y_data, x_title, x_subtitle, color, xticklabels):
    
    ax.plot(x_data, y_data, color=color)
    ax.set_xticklabels(xticklabels, alpha=0.3)
    ax.text(x_title, 1.92, name, fontsize=18, weight='bold', color=color)
    ax.text(x_subtitle, 1.8, period, weight='bold', alpha=0.3)

color = ['#B22222', '#FFA500', '#00BFFF']
presidents = ['BUSH', 'OBAMA', 'TRUMP']
period = ['(2001-2009)', '(2009-2017)', '(2017-2021)']
xticklabels = [['', '2001', '', '2003', '', '2005', '', '2007', '', '2009'],
              ['', '2009', '', '2011', '', '2013', '', '2015', '', '2017'],
              ['2017', '', '2018', '', '2019', '', '2020', '', '2021']]
x_title = [bush['Time'].dt.date.iloc[0] + dt.timedelta(1000), 
           obama['Time'].dt.date.iloc[0] + dt.timedelta(950),
           trump['Time'].dt.date.iloc[0] + dt.timedelta(600)]
x_subtitle = [bush['Time'].dt.date.iloc[0] + dt.timedelta(750),
              obama['Time'].dt.date.iloc[0] + dt.timedelta(800),
              trump['Time'].dt.date.iloc[0] + dt.timedelta(500)]
x_data = [bush['Time'], obama['Time'], trump['Time']]
y_data = [bush['rolling_mean'], obama['rolling_mean'], trump['rolling_mean']]
for i in range(3):
    #Plot the first 3 axes
    #bush: ax1
    #obama: ax2
    #trump: ax3
    plot_presidencies(axes[i],presidents[i], period[i], x_data[i], y_data[i],
                      x_title[i], x_subtitle[i], color[i], xticklabels[i])

    #plot the 4th axis, bush-obama-trump
    ax4.plot(x_data[i], y_data[i],color=color[i])
    ax4.set_xticks([])

#Adding a title and a subtitle
ax1.text(bush['Time'].dt.date.iloc[0]-dt.timedelta(500), 2.35, 'EURO-USD rate averaged 1.22 under the last three US presidents',
         fontsize=20, weight='bold')
ax1.text(bush['Time'].dt.date.iloc[0]-dt.timedelta(500), 2.14, """EURO-USD exchange rates under George W. Bush (2001 - 2009), Barack Obama (2009-2017),
and Donald Trump (2017-2021)""", fontsize=16)

# Adding a signature
ax4.text(bush['Time'].dt.date.iloc[0]-dt.timedelta(600), 0.65, '©Tu_Fei' + ' '*109 + 'Source: European Central Bank',
        color = '#FFFFFF' , backgroundcolor = '#800080',size=14)
plt.show()
    
```
